```python
#draw a line with suitable label in the x axis, y axis and a title.

import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 3 for value in X]
print("Values of X")
print(*range(1,50))
print("Values of Y (thrice of X):")
print(Y)
plt.plot(X, Y)
# Set the x axis label of the current axis.
plt.xlabel("x - axis")
# Set the y axis label of the current axis.
plt.ylabel("y - axis")
# Set a title
plt.title("Draw a line.")
# Display the figure.
plt.show()
```

```python
#draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

import matplotlib.pyplot as plt
with open("test.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph!')
plt.show()
```

```python
#Plot two or more lines with legends, different widths and colors

import matplotlib.pyplot as plt

x1, y1 = [10,20,30], [20,40,10]
x2, y2 = [10,20,30], [40,10,30]

plt.plot(x1, y1, 'b-', linewidth=3, label='line1-width-3')
plt.plot(x2, y2, 'r-', linewidth=5, label='line2-width-5')

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two lines with different widths and colors')

plt.legend()
plt.show()
```

```python
#Plot quantities which have an x and y position

import pylab as pl

x1, y1 = [2, 3, 5, 6, 8], [1, 5, 10, 18, 20]
x2, y2 = [3, 4, 6, 7, 9], [2, 6, 11, 20, 22]

pl.axis([0, 10, 0, 30]) 
pl.plot(x1, y1, 'b*', x2, y2, 'ro')  
pl.show()  
```

```python
#Create plot using subplots(), scatter, and some other methods on the Axes object

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x1, y1 = [2, 3, 5, 6, 8], [1, 5, 10, 18, 20]
x2, y2 = [3, 4, 6, 7, 9], [2, 6, 11, 20, 22]
ax.scatter(x1, y1, color='blue', marker='*', label='Group 1')
ax.scatter(x2, y2, color='red', marker='o', label='Group 2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot with Subplots')
ax.grid(True)
ax.legend()
plt.show()
```

```python
#Generate random data for x and y and create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)  
x = np.random.rand(50)  
y = np.random.rand(50)  

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Random Scatter Plot')
plt.show()
```

```python
#Create a bar chart to compare the sales of different products.You can use fictional data or come up with your own dataset.

import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 420, 300, 500]

# Creating the bar chart
plt.bar(products, sales, color='skyblue')

# Adding labels and title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales Comparison of Different Products')

# Displaying the plot
plt.show()
```

```python
#Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.

import matplotlib.pyplot as plt
import numpy as np

# Generating a random dataset with 1000 data points from a normal distribution
np.random.seed(0)  
data = np.random.normal(size=1000)

# Creating a histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Dataset')

# Displaying the plot
plt.show()
```

```python
#Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.

import matplotlib.pyplot as plt

# Expense categories
categories = ['Housing', 'Transportation', 'Food', 'Utilities', 'Entertainment', 'Healthcare']

# Expense amounts (in dollars)
expenses = [1500, 600, 800, 300, 400, 200]

# Explode the slice with the highest expense
explode = (0.1, 0, 0, 0, 0, 0)  # "Housing" slice will explode out

# Create a pie chart
plt.pie(expenses, labels=categories, explode=explode, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Adding a title
plt.title('Distribution of Expenses in Budget')

# Displaying the plot
plt.show()
```
