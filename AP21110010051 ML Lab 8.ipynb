```python
#Take two 3D coordinates from the user. Find out the distance between these points
#using following distance functions.
#x and y are vectors representing the coordinates of two points.
#Here, k=3
import math

def get_coordinates():
    x1, y1, z1 = map(float, input("Enter the coordinates of the first point (x, y, z): ").split())
    x2, y2, z2 = map(float, input("Enter the coordinates of the second point (x, y, z): ").split())
    return (x1, y1, z1), (x2, y2, z2)

def euclidean_distance(point1, point2):
    x1, y1, z1 = point1
    x2, y2, z2 = point2
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2 + (z2 - z1) ** 2)

def manhattan_distance(point1, point2):
    x1, y1, z1 = point1
    x2, y2, z2 = point2
    return abs(x2 - x1) + abs(y2 - y1) + abs(z2 - z1)

def minkowski_distance(point1, point2, k):
    x1, y1, z1 = point1
    x2, y2, z2 = point2
    return (abs(x2 - x1) ** k + abs(y2 - y1) ** k + abs(z2 - z1) ** k) ** (1 / k)

point1, point2 = get_coordinates()

print("Euclidean distance:", euclidean_distance(point1, point2))
print("Manhattan distance:", manhattan_distance(point1, point2))
print("Minkowski distance (k=3):", minkowski_distance(point1, point2, 3))
```

```python
#Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for
#all the data samples in the feature space. Distance metric is a 2D array, where the
#(i,j)th entry represents the distance between the ith and jth sample points in the feature
#space.
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Calculate the Euclidean Distance Matrix
distance_matrix = np.zeros((X.shape[0], X.shape[0]))

for i in range(X.shape[0]):
    for j in range(i + 1, X.shape[0]):
        distance_matrix[i, j] = distance_matrix[j, i] = np.linalg.norm(X[i] - X[j])

print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
#Import the Iris dataset. Prepare a dataset considering samples belong to any two
#output classes. Draw the scatter plot for all the samples in the new dataset considering
#any two input attributes. Examine the scatter plot to find the equation of a line that
#can separate sample of two classes.
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Select samples belonging to classes 0 and 1
two_class_df = iris_df[iris_df['target'].isin([0, 1])]

# Scatter plot for 'sepal length (cm)' vs 'sepal width (cm)'
plt.figure(figsize=(8, 6))
plt.scatter(two_class_df['sepal length (cm)'], two_class_df['sepal width (cm)'],
            c=two_class_df['target'], cmap='viridis', edgecolor='k', s=100)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.colorbar(label='Target Class')
plt.grid(True)
plt.show()
```
